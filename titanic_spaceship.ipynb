```python
train_y = train_data['Transported']
train_x = train_data.drop(['PassengerId','Destination','Name','Transported'],axis=1)
test_y = test_data.drop(['PassengerId','Destination','Name'],axis=1)
```

```python
def filling_data(train_x):
    train_x['HomePlanet'].fillna(train_x['HomePlanet'].mode()[0], inplace=True)
    train_x['CryoSleep']=train_x['CryoSleep'].fillna(False)
    train_x['VIP'].fillna(train_x['VIP'].mode()[0], inplace=True)
    train_x['Age'].fillna(train_x['Age'].mean(), inplace = True)
    train_x['RoomService']=train_x['RoomService'].fillna(0)
    train_x['FoodCourt']=train_x['FoodCourt'].fillna(0)
    train_x['ShoppingMall']=train_x['ShoppingMall'].fillna(0)
    train_x['Spa']=train_x['Spa'].fillna(0)
    train_x['VRDeck']=train_x['VRDeck'].fillna(0)
    return train_x
train_x = filling_data(train_x)
test_y = filling_data(test_y)
```

```python
train_x['Cabin']=train_x['Cabin'].fillna(method='ffill')
test_y['Cabin']=test_y['Cabin'].fillna(method='ffill')
train_x['deck'] = train_x['Cabin'].apply(lambda x: x.split('/')[0])
train_x['num'] = train_x['Cabin'].apply(lambda x: x.split('/')[1])
train_x['side'] = train_x['Cabin'].apply(lambda x: x.split('/')[2])
test_y['deck'] = test_y['Cabin'].apply(lambda x: x.split('/')[0])
test_y['num'] = test_y['Cabin'].apply(lambda x: x.split('/')[1])
test_y['side'] = test_y['Cabin'].apply(lambda x: x.split('/')[2])
```

```python
del train_x['Cabin']
del test_y['Cabin']
```

```python
f = ['HomePlanet','CryoSleep','VIP','deck','num','side']
l = LabelEncoder()
for i in f:
  train_x[i] = l.fit_transform(train_x[i])
f1 = ['HomePlanet','CryoSleep','VIP','deck','num','side']
l = LabelEncoder()
for i in f1:
  test_y[i] = l.fit_transform(test_y[i])
```

```python
train_x["MoneySpent"] = train_x["RoomService"] + train_x["FoodCourt"] + train_x["ShoppingMall"] + train_x["Spa"] + train_x["VRDeck"] 
test_y["MoneySpent"] = test_y["RoomService"] + test_y["FoodCourt"] + test_y["ShoppingMall"] + test_y["Spa"] + test_y["VRDeck"] 
train_x["Any_Spend"] = train_x["MoneySpent"] > 0
test_y["Any_Spend"] = test_y["MoneySpent"] > 0
train_x["Is_Child"] = train_x["Age"] <= 10
test_y["Is_Child"] = test_y["Age"] <= 10
train_x.head()
train_x.head()
```

```python
test_y.head()
```

```python
train_x["Any_Spend"] = train_x["Any_Spend"].astype(int)
test_y["Any_Spend"] = test_y["Any_Spend"].astype(int)
train_x["Is_Child"] = train_x["Is_Child"].astype(int)
test_y["Is_Child"] = test_y["Is_Child"].astype(int)
```

```python
from sklearn.linear_model import LogisticRegression
r = LogisticRegression()
r.fit(train_x, train_y)
pred_y = r.predict(test_y)
r.score(train_x, train_y)
accuracy = round(r.score(train_x, train_y) * 100, 2)
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=600,max_leaf_nodes=300)
random_forest.fit(train_x, train_y)
pred_y = random_forest.predict(test_y)
random_forest.score(train_x, train_y)
accuracy = round(random_forest.score(train_x, train_y) * 100, 2)
accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(train_x, train_y)
pred_y = knn.predict(test_y)
acc_knn = round(knn.score(train_x, train_y) * 100, 2)
acc_knn
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(train_x, train_y)
pred_y = decision_tree.predict(test_y)
acc_decision_tree = round(decision_tree.score(train_x, train_y) * 100, 2)
acc_decision_tree
```

```python
from xgboost import XGBClassifier
xgb_model = XGBClassifier(gamma = 1.5,
                         subsample = 1.0,
                          max_depth = 5,
                       colsample_bytree = 1.0,
                       n_estimators = 100)
xgb_model = xgb_model.fit(train_x, train_y)
pred_y = xgb_model.predict(test_y)
acc = round(xgb_model.score(train_x, train_y) * 100, 2)
acc
```
